```python
# Scam Detection Dataset Analysis
# This notebook provides basic statistics for each scam detection dataset

import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import os

# Settings for better readability
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 200)
pd.set_option('display.width', 1000)

# Function to print dataset statistics
def analyze_dataset(df, name):
    print(f"\n{'='*50}")
    print(f"Analyzing dataset: {name}")
    print(f"{'='*50}")
    
    # Shape
    print(f"Shape of dataset: {df.shape}")
    
    # First few rows
    print("\nFirst 5 rows:")
    display(df.head(5))
    
    # Column types
    print("\nColumn information:")
    print(df.dtypes)
    
    # Label distribution if present
    for col in ['label', 'labels']:
        if col in df.columns:
            print(f"\nLabel distribution ({col}):")
            label_counts = df[col].value_counts()
            print(label_counts)
            plt.figure(figsize=(8,4))
            label_counts.plot(kind='bar')
            plt.title(f'Label Distribution in {name}')
            plt.xlabel('Label')
            plt.ylabel('Count')
            plt.show()
            break  # avoid double printing if both exist
    
    # Other categorical columns
    for col in ['prediction', 'type', 'personality']:
        if col in df.columns:
            print(f"\n{col.capitalize()} distribution:")
            print(df[col].value_counts())
    
    # Missing values
    print("\nMissing values per column:")
    print(df.isnull().sum())

# Path to Data folder (relative from Code folder)
data_folder = "../Data"

# List all CSV files in Data folder
dataset_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]

print(f"Found {len(dataset_files)} dataset(s): {dataset_files}")

# Analyze each dataset
for file in dataset_files:
    file_path = os.path.join(data_folder, file)
    try:
        df = pd.read_csv(file_path)
        analyze_dataset(df, file)
    except Exception as e:
        print(f"Could not read {file}: {e}")
```
